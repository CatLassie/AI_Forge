# MIDI Player

```python
from pathlib import Path

# https://stackoverflow.com/questions/73924286/play-midi-files-in-google-colab-notebook
from music21 import midi

# https://colab.research.google.com/github/asigalov61/SuperPiano/blob/master/SuperPiano.ipynb#scrollTo=dzrRdTvlJdFd
from midi2audio import FluidSynth
from IPython.display import Audio



# https://stackoverflow.com/questions/73924286/play-midi-files-in-google-colab-notebook
def get_midi_stream(path):
  mf = midi.MidiFile()
  mf.open(path)
  mf.read()
  mf.close()
  s = midi.translate.midiFileToStream(mf)
  return s



# https://colab.research.google.com/github/asigalov61/SuperPiano/blob/master/SuperPiano.ipynb#scrollTo=dzrRdTvlJdFd
# Colab cannot play music directly from music21 - must convert to .wav first
def play_wav(stream):
    out_midi = stream.write('midi')
    out_wav = str(Path(out_midi).with_suffix('.wav'))
    FluidSynth("/content/musicautobot/Yamaha-C5-Salamander-JNv5.1.sf2").midi_to_audio(out_midi, out_wav)
    return Audio(out_wav)
```
